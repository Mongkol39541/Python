```python
import pandas as pd

df = pd.read_csv("EntriesGender.csv")
```

```python
Total = []
for j_1 in range(46):
    M = df["Male"].loc[j_1]
    F = df["Female"].loc[j_1]
    Sum = M+F
    Total.append(Sum)

df["TOTAL"] = pd.DataFrame(Total,index=df.index)
```

```python
Dict = dict(df)
DI = Dict["Discipline"]
DI[46] = "Sum"
DI[47] = "Mean"
```

```python
df = pd.DataFrame(Dict)
```

```python
Discipline_1 = df["Discipline"]
Number = df["Female"]+df["Male"]
Try = ""
Len = 0
Y = 0
Include_Artistic = []
Include_Ball = []
Include_Cyclingl = []
Include_Swimming = []
Include_Canoe = []
Include_Tennis = []
Include_Gymnastics = []
Include_Other = []
ListSigma_Artistic = []
ListSigma_Ball = []
ListSigma_Cyclingl = []
ListSigma_Swimming = []
ListSigma_Canoe = []
ListSigma_Tennis = []
ListSigma_Gymnastics = []
ListSigma_Other = []

for DIS in range(46):
    Try = Discipline_1[DIS]
    Len = Number[DIS]
    for Artistic in range(20):
        X = Try[Artistic:Artistic+8]
        if X == "Artistic":
            Include_Artistic.append(Try)
            ListSigma_Artistic.append(Len)
            Y = 1
            X = 0
            break
    if X != 0 :
        ListSigma_Artistic.append(0)
    for ball in range(20):
        X = Try[ball:ball+4]
        if X == "ball":
            Include_Ball.append(Try)
            ListSigma_Ball.append(Len)
            Y = 1
            X = 0
            break
    if X != 0 :
        ListSigma_Ball.append(0)
    for Cycling in range(20):
        X = Try[Cycling:Cycling+7]
        if X == "Cycling":
            Include_Cyclingl.append(Try)
            ListSigma_Cyclingl.append(Len)
            Y = 1
            X = 0
            break
    if X != 0 :
        ListSigma_Cyclingl.append(0)
    for Swimming in range(20):
        X = Try[Swimming:Swimming+8]
        if X == "Swimming":
            Include_Swimming.append(Try)
            ListSigma_Swimming.append(Len)
            Y = 1
            X = 0
            break
    if X != 0 :
        ListSigma_Swimming.append(0)
    for Canoe in range(20):
        X = Try[Canoe:Canoe+5]
        if X == "Canoe":
            Include_Canoe.append(Try)
            ListSigma_Canoe.append(Len)
            Y = 1
            X = 0
            break
    if X != 0 :
        ListSigma_Canoe.append(0)
    for Tennis in range(20):
        X = Try[Tennis:Tennis+6]
        if X == "Tennis":
            Include_Tennis.append(Try)
            ListSigma_Tennis.append(Len)
            Y = 1
            X = 0
            break
    if X != 0 :
        ListSigma_Tennis.append(0)
    for Gymnastics in range(20):
        X = Try[Gymnastics:Gymnastics+10]
        if X == "Gymnastics":
            Include_Gymnastics.append(Try)
            ListSigma_Gymnastics.append(Len)
            Y = 1
            X = 0
            break
    if X != 0 :
        ListSigma_Gymnastics.append(0)
    if Y == 0:
        Include_Other.append(Try)
        ListSigma_Other.append(Len)
    if Y != 0:
        ListSigma_Other.append(0)
    Try = ""
    Len = 0
    Y = 0
```

```python
ListSigma_Female = list(df["Female"])[0:46]
ListSigma_Man = list(df["Male"])[0:46]
```

```python
List = [
    ListSigma_Artistic,
    ListSigma_Ball,
    ListSigma_Cyclingl,
    ListSigma_Swimming,
    ListSigma_Canoe,
    ListSigma_Tennis,
    ListSigma_Gymnastics,
    ListSigma_Other,
    ListSigma_Female,
    ListSigma_Man
    ]
```

```python
import numpy as np
ar = np.array(List)
SI = []
ME = []
for calculate in range(len(List)):
    S = ar[calculate].sum()
    M_1 = ar[calculate].mean()
    M_2 = round(M_1,1)
    SI.append(S)
    ME.append(M_2)
```

```python
for Match in range(10):
    List[Match].append(SI[Match])
    List[Match].append(ME[Match])

DI_New = {
    "Artistic":ListSigma_Artistic,
    "Ball":ListSigma_Ball,
    "Cycling":ListSigma_Cyclingl,
    "Swimming":ListSigma_Swimming,
    "Canoe":ListSigma_Canoe,
    "Tennis":ListSigma_Tennis,
    "Gymnastics":ListSigma_Gymnastics,
    "Other":ListSigma_Other,
    "Female":ListSigma_Female,
    "Man":ListSigma_Man
    }
Data = pd.DataFrame(DI_New,index=df["Discipline"])
pd.set_option("display.max_row",Data.shape[0])
pd.set_option("display.max_column",Data.shape[0])
Data
```
