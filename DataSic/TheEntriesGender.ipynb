```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("EntriesGender.csv")
df
```

```python
Discipline = df["Discipline"]
Female = df["Female"]
Male = df["Male"]
Sum = 0
TheSum = []

for sum_1 in range(46):
    Sum = Female[sum_1]+Male[sum_1]
    TheSum.append(Sum)
    Sum = 0
```

```python
Ar = np.array([list(Female),list(Male)])
SunFemale = Ar[0].sum()
SunMale = Ar[1].sum()
```

```python
from tkinter import*
from tkinter import ttk
Window = Tk()
Window.title("EntriesGender")
Window.geometry("1200x720")
LB = Label(Window,text="มีนักกีฬาแต่ละประเภทกี่คน",font=("Arial Bold",20))
LB.grid(column=0,row=0)
LB1 = Label(Window,text="มีนักกีฬาหญิงทั้งหมด"+str(SunFemale)+"คน",font=("Arial Bold",20))
LB1.grid(column=1,row=1)
LB2 = Label(Window,text="มีนักกีฬาชายทั้งหมด"+str(SunMale)+"คน",font=("Arial Bold",20))
LB2.grid(column=1,row=2)

Tex = ttk.Combobox(Window,values = list(Discipline),width=20)
Tex.grid(column=0,row=2)

def BUTTON1():
    TEXT = Tex.get()
    for i in range(len(list(Discipline))):
        if str(TEXT) == str(list(Discipline)[i]):
            LB.configure(text="มีนักกีฬาประเภท"+TEXT+"เท่ากับ"+str(TheSum[i])+"คน")

BT = Button(Window,text="Click",font=("Arial Bold",20),command=BUTTON1)
BT.grid(column=0,row=1)
Window.mainloop()
```
