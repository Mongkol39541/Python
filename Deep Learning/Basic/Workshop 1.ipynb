```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot('sepal_length', 'sepal_width', data=df, hue='species')
plt.show()
```

```python
sns.scatterplot('petal_length', 'petal_width', data=df, hue='species')
plt.show()
```

```python
sns.pairplot(df, hue='species', height=2.0)
plt.show()
```

```python
columns = ['sepal_length', 'sepal_width', 'petal_length', 'species']
sns.pairplot(df[columns], hue='species', height=2.0)
plt.show()
```
