```python
def linear05(x):
    result = 0.5 * x
    return result

def linear1(x):
    result = x
    return result

def linear2(x):
    result = 2.0 * x
    return result

linear05(3)
```

```python
linear1(3)
```

```python
linear2(3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

val = np.arange(-7, 7, 0.15)
out05 = [linear05(x) for x in val]
out1 = [linear1(x) for x in val]
out2 = [linear2(x) for x in val]

plt.plot(val, out05, lw=2., c='g')
plt.plot(val, out1, lw=2., c='m')
plt.plot(val, out2, lw=2., c='b')
plt.ylabel('y (Output)')
plt.xlabel('x (input)')
plt.axhline(y = 0.0, color = 'r', linestyle = '--', lw=1.2)
plt.grid()
plt.show()
```
