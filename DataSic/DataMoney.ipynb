```python
import pandas as pd
import matplotlib as plt
data_medal = pd.read_csv("medals.csv")
```

```python
Team = data_medal["Team/NOC"]
Go = data_medal["Gold"]
Si = data_medal["Silver"]
Br = data_medal["Bronze"]
Total = []
SuperGo = 0
SuperSi = 0
SuperBr = 0
SuperSum = 0

for mon in range(93):
    Sum = Go.loc[mon] + Si.loc[mon] + Br.loc[mon]
    Total.append(Sum)
    SuperGo += Go.loc[mon]
    SuperSi += Si.loc[mon]
    SuperBr += Br.loc[mon]
    SuperSum += Sum

data_medal["SUM"] = pd.DataFrame(Total,index=data_medal.index)
```

```python
Dict = dict(data_medal)

ALR = Dict["Rank"]
ALR[93] = "-"

DI = Dict["Team/NOC"]
DI[93] = "AllTeam"

ALG = Dict["Gold"]
ALG[93] = SuperGo

ALS = Dict["Silver"]
ALS[93] = SuperSi

ALB = Dict["Bronze"]
ALB[93] = SuperBr

AL = Dict["SUM"]
AL[93] = SuperSum
```

```python
data_medal = pd.DataFrame(Dict)
pd.set_option("display.max_rows",data_medal.shape[0])
data_medal.drop(["Rank"],axis=1,inplace=True)
data_medal
```
