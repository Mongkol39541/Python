```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
graph = pd.read_csv("n_movies.csv", encoding='latin')
graph
```

Number Of Movies Of Each Genre

```python
genre = []
set_genre = []
for item in graph["genre"]:
    try:
        item = item.split(', ')
        for txt in item:
            set_genre.append(txt)
        genre.append(item)
    except:
        genre.append(["None"])
set_genre = set(set_genre)
set_genre = list(set_genre)
graph["genre"] = genre
graph
```

```python
set_cout = []
for att in set_genre:
    cout = 0
    for che in genre:
        if att in che:
            cout += 1
    set_cout.append(cout)
```

```python
plt.figure(figsize=(8, 8))
x1 = set_genre
y1 = set_cout
x_pos = np.arange(len(x1))
plt.xticks(x_pos, x1, rotation=90)
plt.bar(x1, y1)
plt.title(label= "Number Of Movies Of Each Genre", fontsize=18, color="black")
plt.xlabel('Genre')
plt.ylabel('Count of genre')
plt.show()
```

Hunter X Hunter

```python
plt.figure(figsize=(8, 8))
x1 =set_genre
y1 = set_cout
plt.pie(y1, labels = x1, wedgeprops = { 'linewidth' : 2, 'edgecolor' : 'white'})
plt.title(label= "Hunter X Hunter", fontsize=18, color="black")
plt.show()
```

Duration vs Rating vs Votes

```python
cout_min = []
for item in graph["duration"]:
    try:
        item = item.split(" min")
        cout_min.append(int(item[0]))
    except:
        item = np.random.randint(60, 120)
        cout_min.append(item)
graph["duration"] = cout_min
graph
```

```python
plt.figure(figsize=(8, 8))
ax = plt.axes(projection ='3d')
ax.scatter(graph["duration"], graph["rating"], graph["votes"], c= graph["votes"])
ax.set_title('Duration vs Rating vs Votes', fontsize=24)
ax.set_xlabel('Duration', fontsize=18)
ax.set_ylabel('Rating', fontsize=18)
ax.set_zlabel('Votes', fontsize=18)
plt.show()
```
