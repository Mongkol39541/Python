```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("medals.csv")
df
```

```python
Team = df["Team/NOC"]
Gold = df["Gold"]
Silver = df["Silver"]
Bronze = df["Bronze"]
Sum = 0
TheSum = []

for i_1 in range(len(Team)):
    Sum = Gold[i_1]+Silver[i_1]+Bronze[i_1]
    TheSum.append(Sum)
    Sum = 0

AllSum = np.array(TheSum)
Total = AllSum.sum()
```

```python
from tkinter import*
from tkinter import ttk
Window = Tk()
Window.title("EntriesGender")
Window.geometry("1200x720")
LB = Label(Window,text="แต่ละประเทศมีเหรียญทั้งหมดเท่าไหร่",font=("Arial Bold",20))
LB.grid(column=0,row=0)
LB1 = Label(Window,text="มีเหรียญทั้งหมดเท่ากับ"+str(Total)+"เหรียญ",font=("Arial Bold",20))
LB1.grid(column=1,row=1)

Tex = ttk.Combobox(Window,values = list(Team),width=20)
Tex.grid(column=0,row=2)

def BUTTON1():
    TEXT = Tex.get()
    for i in range(len(list(Team))):
        if str(TEXT) == str(list(Team)[i]):
            LB.configure(text="ประเทศ"+TEXT+"มีเหรียญเท่ากับ"+str(TheSum[i])+"เหรียญ")

BT = Button(Window,text="Click",font=("Arial Bold",20),command=BUTTON1)
BT.grid(column=0,row=1)
Window.mainloop()
```
