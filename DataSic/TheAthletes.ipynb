```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("athletes.csv")
df
```

```python
LS = []
SuperLS = []
SUM = 0

for i_1 in range(len(df["NOC"])):
    Noc0 = df["NOC"].loc[i_1]
    LS.append(Noc0)
Noc1 = list(set(LS))
```

```python
for i_1_1 in range(len(Noc1)):
    for i_2 in range(len(df["NOC"])):
        Noc2 = df["NOC"].loc[i_2]
        if Noc1[i_1_1] == Noc2 :
            SUM += 1
    SuperLS.append(SUM)
    SUM = 0
```

```python
List = [Noc1,SuperLS]
```
