```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("netflix_titles.csv")
df
```

```python
List = []
Ty = ""
Y = 0
for i in range(7787):
    Ty = df["country"].loc[i]
    if type(df["country"].loc[i]) != type(1.0):
        S = df["country"].loc[i]
        for j in range(100):
            X = Ty[j:j+1]
            if X == ",":
                Y = 1
        if Y == 0 :
            List.append(Ty)
        Y = 0
    Set = set(List)
```

```python
COU_0 = list(Set)
TS_1 = df["type"]
CY_1 = df["country"]
SuperList1 = []
SuperList2 = []
XXT = 0
XXM = 0
Keep = []

for Check_0 in range(len(COU_0)):
    for Check_1 in range(7787):
        if COU_0[Check_0] == CY_1.loc[Check_1]:
            SuperList1.append(TS_1.loc[Check_0])
            SuperList2.append(CY_1.loc[Check_0])
            if TS_1[Check_1] == "TV Show":
                XXT += 1
            if TS_1[Check_1] == "Movie":
                XXM += 1
    Keep.append([XXT,XXM])
    XXT = 0
    XXM = 0

KeepTV = []
KeepMO = []

for Check_2 in range(len(Keep)):
    KT = Keep[Check_2][0]
    KM = Keep[Check_2][1]
    KeepTV.append(KT)
    KeepMO.append(KM)

Dict = {
    "TV Show":KeepTV,
    "Movie":KeepMO
    }
```

```python
DF = pd.DataFrame(Dict,index=COU_0)
pd.set_option("display.max_row",DF.shape[0])

for Check_3 in range(len(Keep)):
    AR = np.array([KeepTV[Check_3],KeepMO[Check_3]])
    NameAR = ["TV Show","Movie"]
    myexplode = [0.2, 0]
    font1 = {'family':'serif','color':'white','size':20}
    plt.title(COU_0[Check_3],fontdict = font1)
    plt.pie(AR,labels = NameAR,explode = myexplode,shadow = True)
    plt.show()
```
